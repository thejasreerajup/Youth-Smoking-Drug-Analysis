# FEATURE ENGINEERING

Feature Engineering is an important step because it helps convert raw data into meaningful information that a model can easily understand. By creating new features from existing data, such as combining related values or transforming data into a simpler form, important patterns in the dataset become clearer. This process improves data quality, reduces unnecessary noise, and helps the machine learning model produce more accurate and reliable results.

### Numeric Year (for trend analysis)

```python
df['Year_Num'] = df['Year'].dt.year
df
```

### Social Influence Score 

```python
df['Social_Influence_Score'] = (
    df['Peer_Influence'] +
    df['Media_Influence']
)
df
```

### Family Support Score

```python

df['Family_Support_Score'] = (
    df['Family_Background'] +
    df['Parental_Supervision'] +
    df['Community_Support']
)
df
```

### Mental & Support Score 

```python
df['Mental_Support_Score'] = (
    df['Mental_Health'] +
    df['Access_to_Counseling'].map({'Yes': 1, 'No': 0})
)
df
```

### Awareness Score

```python
df['Awareness_Score'] = (
    df['Substance_Education'].map({'Yes': 1, 'No': 0}) +
    df['School_Programs'].map({'Yes': 1, 'No': 0})
)
df
```

### Overall Risk Score 

```python
df['Overall_Risk_Score'] = (
    df['Social_Influence_Score']
    - df['Family_Support_Score']
    - df['Awareness_Score']
)
df
```

# RELATIONSHIP AND TREND ANALYSIS

Relationship and trend analysis was used to study how factors such as age group, gender, peer influence, mental health, and access to counseling are related to smoking prevalence and drug experimentation. ### Does higher peer influence lead to higher smoking prevalence?

```python
high_peer = df[df['Peer_Influence'] > 7]['Smoking_Prevalence'].mean()
low_peer = df[df['Peer_Influence'] <= 7]['Smoking_Prevalence'].mean()

print("High peer influence – Avg Smoking:", high_peer)
print("Low peer influence – Avg Smoking:", low_peer)
```

 # VISUALISATION

Data visualization techniques were applied to analyze patterns and trends in youth smoking and drug experimentation. Visual representations such as charts and graphs helped identify age group differences, gender-based variations, time-based trends, and relationships between social, psychological, and behavioral factors.

### Smoking by Age Group

```python
age_smoking = df.groupby('Age_Group')['Smoking_Prevalence'].mean()

age_smoking.plot(kind='bar')
plt.title('Smoking Prevalence by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Smoking Prevalence')
plt.show()
```

### Drug Experimentation by Age Group

```python
age_drug = df.groupby('Age_Group')['Drug_Experimentation'].mean()

age_drug.plot(kind='bar')
plt.title('Drug Experimentation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Drug Experimentation')
plt.tight_layout()
plt.show()
```

### Peer Influence vs Smoking 

```python
peer_avg = df.groupby('Peer_Influence')['Smoking_Prevalence'].mean().reset_index()
plt.plot(peer_avg['Peer_Influence'],
         peer_avg['Smoking_Prevalence'],
         marker='o')

plt.title('Peer Influence vs Smoking Prevalence')
plt.xlabel('Peer Influence Level')
plt.ylabel('Average Smoking Prevalence')
plt.show()
```

### Socioeconomic Status vs Smoking

```python
bp = plt.boxplot(
    [df[df['Socioeconomic_Status'] == status]['Smoking_Prevalence']
     for status in df['Socioeconomic_Status'].unique()],
    labels=df['Socioeconomic_Status'].unique(),
    patch_artist=True
)

colors = ['#90DBF4', '#A3C4F3', '#CFBAF0'] 


for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

plt.title('Smoking Prevalence by Socioeconomic Status')
plt.xlabel('Socioeconomic Status')
plt.ylabel('Smoking Prevalence')
plt.show()
```

### Access to Counseling vs Drug Experimentation

```python
counseling_drug = df.groupby('Access_to_Counseling')['Drug_Experimentation'].mean()

counseling_drug.plot(kind='bar')
plt.title('Access to Counseling vs Drug Experimentation')
plt.xlabel('Access to Counseling')
plt.ylabel('Drug Experimentation')
plt.show()
```

###     Distribution of Smoking Prevalence

```python
counts, bins, patches = plt.hist(df['Smoking_Prevalence'], bins=8, color='skyblue', edgecolor='black')
for count, patch in zip(counts, patches):
    plt.text(patch.get_x() + patch.get_width()/2, count, int(count), ha='center', va='bottom')

plt.title('Smoking Prevalence Distribution')
plt.xlabel('Smoking Prevalence')
plt.ylabel('Count')
plt.show()
```

### Smoking vs Drug Experimentation by Gender 

```python
df.groupby('Gender')[['Smoking_Prevalence','Drug_Experimentation']].mean().plot(kind='bar')
plt.title('Smoking vs Drug Experimentation by Gender')
plt.xlabel('Gender')
plt.ylabel('Rate')
plt.show()
```

### Average Smoking Prevalence by Media Influence

```python
media_grouped = df.groupby('Media_Influence')['Smoking_Prevalence'].mean()

plt.pie(
    media_grouped.values,
    labels=media_grouped.index,
    autopct='%1.1f%%',
    colors=['skyblue', 'orange', 'green', 'red', 'purple'],  # optional
    startangle=90
)
plt.title('Average Smoking Prevalence by Media Influence')
plt.show()
```

### Smoking Prevalence Trend Over Years

```python
yearly_smoking = df.groupby('Year')['Smoking_Prevalence'].mean()

plt.figure(figsize=(8,5))
plt.plot(yearly_smoking.index, yearly_smoking.values, marker='o')

plt.title("Smoking Prevalence Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Average Smoking Prevalence")

plt.xticks(yearly_smoking.index)

plt.tight_layout()
plt.show()
```

### Year-wise Smoking Trend by Gender

```python
for g in df['Gender'].unique():
    yearly_gender = df[df['Gender'] == g].groupby('Year')['Smoking_Prevalence'].mean()
    plt.plot(yearly_gender.index, yearly_gender.values, marker='o', label=g)

plt.title("Year-wise Smoking Trend by Gender")
plt.xlabel("Year")
plt.ylabel("Average Smoking Prevalence")
plt.legend()

plt.xticks(sorted(df['Year'].unique()))

plt.tight_layout()
plt.show()
```

### Drug Experimentation Trend over Years

```python
yearly_drug = df.groupby('Year')['Drug_Experimentation'].mean()

plt.figure(figsize=(8,5))
plt.plot(yearly_drug.index, yearly_drug.values, marker='o')

plt.title("Drug Experimentation Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Average Drug Experimentation")

plt.xticks(yearly_drug.index)

plt.tight_layout()
plt.show()
```

### Year-wise Drug Experimentation by Gender

```python
gender_year_drug = df.groupby(['Year', 'Gender'])['Drug_Experimentation'].mean().unstack()
gender_year_drug.plot(kind='bar', figsize=(9,5))
plt.title("Year-wise Drug Experimentation by Gender")
plt.xlabel("Year")
plt.ylabel("Average Drug Experimentation")
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

### Gender Distribution

```python
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

### Smoking by Age Group

```python
groups = df['Age_Group'].unique()
data = [df[df['Age_Group']==g]['Smoking_Prevalence'] for g in groups]
plt.violinplot(data, showmeans=True, showmedians=True)
plt.xticks(range(1,len(groups)+1), groups, rotation=45)
plt.title(" Smoking by Age Group")
plt.tight_layout()
plt.show()
```

### Smoking vs Drug Experimentation by Age Group 

```python
df.groupby('Age_Group')[['Smoking_Prevalence','Drug_Experimentation']].mean().plot(kind='bar')
plt.title("Smoking vs Drug Experimentation by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Value")
plt.tight_layout()
plt.show()
```

### Distribution of Smoking, Drug Experimentation & Mental Health

```python
plt.figure(figsize=(10,6))
bins = np.linspace(0, 100, 11)
plt.hist(df['Smoking_Prevalence'], bins=bins, alpha=0.7, label='Smoking Prevalence', color='blue', rwidth=0.3)
plt.hist(df['Drug_Experimentation'], bins=bins, alpha=0.7, label='Drug Experimentation', color='red', rwidth=0.3)
plt.hist(df['Mental_Health'], bins=bins, alpha=0.7, label='Mental Health', color='green', rwidth=0.3)

plt.title('Distribution of Smoking, Drug Experimentation & Mental Health')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```
